```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
conditions = pd.read_csv('csv/conditions.csv')

# Count the occurrences of each DESCRIPTION and select the top 10
top_conditions = conditions['DESCRIPTION'].value_counts().head(10)

# Plotting the top 10 conditions
plt.figure(figsize=(10, 6))
top_conditions.plot(kind='bar')
plt.title('Top 10 Conditions')
plt.xlabel('Condition')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
