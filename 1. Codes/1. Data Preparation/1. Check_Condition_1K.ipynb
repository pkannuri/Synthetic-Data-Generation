```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
conditions = pd.read_csv('csv_filtered/conditions.csv')

# Filter rows where DESCRIPTION contains 'disorder'
disorder_conditions = conditions[conditions['DESCRIPTION'].str.contains('disorder', case=False)]

# Count the occurrences of each DESCRIPTION in the filtered DataFrame and select the top 10
top_disorders = disorder_conditions['DESCRIPTION'].value_counts().head(10)

# Plotting the top 10 disorders
plt.figure(figsize=(10, 6))
top_disorders.plot(kind='bar')
plt.title('Top 10 Disorders')
plt.xlabel('Disorder')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
