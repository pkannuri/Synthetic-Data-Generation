```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('merged_csv/merged_df.csv')
df
```

# 1. Patient Information

```python
# Step 1: Remove rows where 'DEATHDATE' is not NaN
df1 = df[pd.isna(df['DEATHDATE'])]

# Step 2: Remove specified columns
columns_to_remove = [
    'DEATHDATE', 'DRIVERS', 'PASSPORT', 'PREFIX', 'FIRST', 'LAST', 'SUFFIX', 
    'MAIDEN', 'BIRTHPLACE', 'ADDRESS', 'CITY'
]
df2 = df1.drop(columns=columns_to_remove)
df2
```

# 2. Allergies

```python
# Identify the start and end column indexes for allergy attributes
allergy_start = df2.columns.get_loc("Aspirin")
allergy_end = df2.columns.get_loc("Tree nut (substance)") + 1

# Isolate allergy columns
allergy_data = df2.iloc[:, allergy_start:allergy_end]

# Sum occurrences of each allergy and sort
allergy_occurrences = allergy_data.sum().sort_values(ascending=False)

# Select the top 10 most frequent allergies
top_10_allergies = allergy_occurrences.head(10)
top_10_allergies_names = top_10_allergies.index.tolist()
```

```python
top_10_allergies_names
```

```python
# Choose 5 allergies relevant to 'Viral sinusitis'

top_5_allergies_names = ['Mold (organism)', 
 'Animal dander (substance)',
 'Grass pollen (substance)',
 'House dust mite (organism)',
 'Tree pollen (substance)']
```

```python
# Mold (organism): Mold spores can easily be inhaled and cause irritation and inflammation in the nasal passages and sinuses, potentially leading to sinusitis.
# House dust mite (organism): Dust mites are a common trigger for allergic rhinitis, which can inflame the sinuses and contribute to the development of sinusitis.
# Grass pollen (substance): Grass pollen is a well-known trigger for allergic rhinitis, especially during high pollen seasons, leading to nasal inflammation that could block sinuses.
# Animal dander (substance): Pet dander is a common allergen that can cause persistent allergic rhinitis, contributing to chronic sinus issues.
# Tree pollen (substance): Similar to grass pollen, tree pollen is a significant trigger for seasonal allergies that can inflame the nasal passages and lead to sinusitis.
```

```python
top_5_allergies_names
```

```python
# Combine non-allergy data with the top 5 allergy columns
non_allergy_columns = (df2.columns[:allergy_start].tolist() + 
                       df2.columns[allergy_end:].tolist())

non_allergy_columns.remove('allergy_DATE')


# Ensure 'allergy_DATE' is at the front of the allergies data
allergies_with_date = ['allergy_DATE'] + top_5_allergies_names

# Combine non-allergy data with the 'allergy_DATE' and top 5 allergy columns
df3 = pd.concat([df2[non_allergy_columns], df2[allergies_with_date]], axis=1)
df3
```

# 3. Careplans

```python
# Convert 'careplan_START' and 'careplan_STOP' columns to datetime format
df3['careplan_START'] = pd.to_datetime(df3['careplan_START'], errors='coerce')
df3['careplan_STOP'] = pd.to_datetime(df3['careplan_STOP'], errors='coerce')

# Find the latest date in the dataset
latest_date = df3[['careplan_START', 'careplan_STOP']].max().max()
latest_date
```

```python
# Calculate the criteria date for the past 24 months
criteria_date = latest_date - pd.DateOffset(months=24)
criteria_date
```

```python
# Define a function to determine if a careplan exists within the last 24 months
def is_careplan_within_24(start_date, stop_date, criteria_date):
    if pd.isnull(start_date) and pd.isnull(stop_date):
        # If both start and stop dates are NaN, return NaN
        return np.nan
    else:
        # If either start or stop date is within the past 24 months, return 1, else return 0
        return int((start_date >= criteria_date) | (stop_date >= criteria_date))

# Apply the function across the dataframe to create the 'careplan_within_24' column
df3['careplan_within_24'] = df3.apply(lambda x: is_careplan_within_24(x['careplan_START'], x['careplan_STOP'], criteria_date), axis=1)

# Display the first few rows to verify the new column
print(df3[['careplan_START', 'careplan_STOP', 'careplan_within_24']].head())
```

```python
# Calculate the length of careplan for each patient (in days)
df3['careplan_length'] = (df3['careplan_STOP'] - df3['careplan_START']).dt.days
```

```python
# Assuming df3 already contains 'careplan_within_24' and 'careplan_length' columns

# Identify indices for careplan data range
careplan_start_index = df3.columns.get_loc('careplan_Id')
careplan_end_index = df3.columns.get_loc('careplan_REASONDESCRIPTION') + 1

# Columns before careplan data
columns_before_careplan = df3.columns[:careplan_start_index]

# Columns after careplan data, excluding 'careplan_within_24' and 'careplan_length'
columns_after_careplan = [col for col in df3.columns[careplan_end_index:] if col not in ['careplan_within_24', 'careplan_length']]

# Careplan data columns, including 'careplan_within_24' and 'careplan_length' placed before the careplan-specific columns
careplan_columns_with_new = list(df3.columns[careplan_start_index:careplan_end_index]) + ['careplan_within_24', 'careplan_length']

# Combine all parts in the new order
new_order = list(columns_before_careplan) + list(columns_after_careplan) + careplan_columns_with_new

# Reorder DataFrame columns
df4 = df3[new_order]
df4
```

# 4. Encounters

```python
# Function to determine encounter type
def determine_encounter_type(description):
    if description == 'Outpatient procedure':
        return 'outpatient'
    elif description == 'Emergency room admission (procedure)':
        return 'emergency room visit'
    else:
        return 'inpatient'

# Apply the function to the 'encounter_DESCRIPTION' column to create 'encounter_type'
df4['encounter_type'] = df4['encounter_DESCRIPTION'].apply(determine_encounter_type)

# Assuming 'encounter_REASONDESCRIPTION' is the column after which you want to insert 'encounter_type'
# Find the position of 'encounter_REASONDESCRIPTION'
position = df4.columns.get_loc('encounter_REASONDESCRIPTION') + 1

# Create a list of column names in the desired order
columns = list(df4.columns)
columns.insert(position, columns.pop(columns.index('encounter_type')))

# Reorder the DataFrame according to the new column order
df5 = df4[columns]
df5
```
