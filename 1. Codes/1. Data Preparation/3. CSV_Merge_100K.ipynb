```python
import pandas as pd

allergies = pd.read_csv('converted_csv/allergies_converted_new.csv')
careplans = pd.read_csv('converted_csv/careplans_converted_new.csv')
conditions = pd.read_csv('converted_csv/conditions_converted_new.csv')
encounters = pd.read_csv('converted_csv/encounters_converted_new.csv')
immunizations = pd.read_csv('converted_csv/immunizations_converted_new.csv')
medications = pd.read_csv('converted_csv/medications_converted_new.csv')
observations = pd.read_csv('converted_csv/observations_converted_new.csv')
patients = pd.read_csv('converted_csv/patients_converted_new.csv')
procedures = pd.read_csv('converted_csv/procedures_converted_new.csv')

print("Shape of allergies DataFrame:", allergies.shape)
print("Shape of careplans DataFrame:", careplans.shape)
print("Shape of conditions DataFrame:", conditions.shape)
print("Shape of encounters DataFrame:", encounters.shape)
print("Shape of immunizations DataFrame:", immunizations.shape)
print("Shape of medications DataFrame:", medications.shape)
print("Shape of observations DataFrame:", observations.shape)
print("Shape of patients DataFrame:", patients.shape)
print("Shape of procedures DataFrame:", procedures.shape)
```

# Merge on PATIENT

```python
# List of all DataFrames for easy iteration
dataframes = [allergies, careplans, conditions, encounters, immunizations, medications, observations, procedures]

# Remove 'Unnamed: 0' columns if they exist in any DataFrame
for df in dataframes:
    if 'Unnamed: 0' in df.columns:
        df.drop('Unnamed: 0', axis=1, inplace=True)

# Merge all DataFrames on 'PATIENT', starting with 'patients'
merged_df = patients.copy()  # Starting point for merging
for df in dataframes:
    merged_df = pd.merge(merged_df, df, on='PATIENT', how='left')

merged_df
```

```python
merged_df.to_csv('merged_csv/merged_df_100K.csv', index=False)
```
